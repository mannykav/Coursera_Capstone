```python
#Create a handle, page, to handle the contents of the website
#page = requests.get(url)
URL = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
page = requests.get(URL)
```

```python
import lxml.html as lh
#Store the contents of the website under doc
doc = lh.fromstring(page.content)
```

```python
#Parse data that are stored between <tr>..</tr> of HTML
tr_elements = doc.xpath('//tr')
```

```python
[len(T) for T in tr_elements[:12]]
```

Pull website data i.e. //tr for table information. Loop through each element to get header data and 
following rows. 

```python
tr_elements = doc.xpath('//tr')
#Create empty list
col=[]
i=0   
    #For each row, store each first element (header) and an empty list
for t in tr_elements[0]:
    i+=1
    name=t.text_content()

    print (i,name)
    col.append((name,[]))
```

```python
#Since out first row is the header, data is stored on the second row onwards
for j in range(1,len(tr_elements)):
    #T is our j'th row
    T=tr_elements[j]
    
    #If row is not of size 3, the //tr data is not from our table 
    if len(T)!=3:
        break
    
    #i is the index of our column
    i=0
    
    #Iterate through each element of the row
    for t in T.iterchildren():
        data=t.text_content() 
        #Check if row is empty
        if i>0:
        #Convert any numerical value to integers
            try:
                data=int(data)
            except:
                pass
        #Append the data to the empty list of the i'th column
        col[i][1].append(data)
        #Increment i for the next column
        i+=1
```

```python
[len(C) for (title,C) in col]
```

```python
Dict={title:column for (title,column) in col}
df=pd.DataFrame(Dict)
```

```python
df.head()
```

Remove \n from columns and column headings.

```python
df2= df.replace(r'\n',  '', regex=True)
df2.head()
```

```python
df2.columns = df.columns.str.replace("\n", "")
list(df2)
```

Remove those with Not Assigned. 

```python
# df3=df2[df2.Borough !="Not assigned"]
df3=df2[~df2.Borough.str.contains("Not assigned")]
df3.head()
```

```python
df4 = df3.reset_index(drop=True)
# del df4['index']
# del df4['level_0']
df4.head(10)
```

```python
df5= df4.replace('/',',',regex=True)
df5.head(10)
```

```python
import folium # map rendering library
```

```python
address = 'Toronto, Canada'

geolocator = Nominatim(user_agent="to_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto are {}, {}.'.format(latitude, longitude))
```

```python
# create map of New York using latitude and longitude values
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, borough, neighborhood in zip(neighborhoods['Latitude'], neighborhoods['Longitude'], neighborhoods['Borough'], neighborhoods['Neighborhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```
